# Applying groupby operation on original_language column and for each will se the dataset

```python
group_language = imdb_clean.groupby('original_language')
#It will return a groupby Object
```

#Let's print all the language name one by one with the dataframe where original_language will be same

```python
total_rows = 0
dict_lang_count={}
for language , df_grp_language in group_language:
  print("*"*25,language,'*'*25)
  print(df_grp_language)
  total_rows += df_grp_language.shape[0]
  dict_lang_count[language] = df_grp_language.shape[0]
  print('-'*100)
```
